### Named Entity Recognition Using Spacy - Article From a Columnist

```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
import re
import spacy
from spacy import displacy
NER=spacy.load('en_core_web_sm')
```

```python
page=requests.get('https://www.karar.com/yazarlar/zafer-acar/astronot-001-1589950')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
soup.prettify
```

```python
df=soup.find('article',{'class':'article-content'}).text
```

```python
df
```

```python
from textblob import TextBlob
```

```python
TextBlob(df).sentiment.polarity
```

```python
text=NER(df)
```

```python
for word in text.ents:
    print(word.text,word.label_)
```

```python
displacy.render(text)
```

```python
displacy.render(text,style='ent',jupyter=True)
```

```python
spacy.explain('CARDINAL')
```
